#### Curve fitting and derivatives

This example was inspired by one of the homework in the MSDS 400, Week 6 exercises.

A family has recently built a dam on their property to create a small lake.  They want to know if the dam is high enough to hold water during the rainy season.

```python
import numpy as np
import matplotlib.pyplot as plt
```

The weekly rainfall figures throughout the year are approximately as follows:

```python
weekly_rainfall = [12, 12, 12, 12, 12, 12, 12, 12, 18, 21, 27, 29, 40, 34, 38, 39, 42, 46, 54, 58, 56, 60, 60, 68, 
                   70, 72, 65, 60, 64, 65, 51, 54, 44, 43, 36, 36, 33, 31, 26, 25, 20, 20, 15, 16, 18, 12, 12, 12, 
                   12, 12, 12, 12]
```

```python
weeks = [i+1 for i in range(0, len(weekly_rainfall))]
rainy_season_weeks = weeks[6:46]
weekly_rainfall_season = weekly_rainfall[6:46]
```

```python
plt.title('Rainfall by Week for Small Town, USA')
plt.xticks(np.arange(10, 50, 5))
plt.xlabel('Weeks')
plt.ylabel('Rainfall in cm')
plt.plot(rainy_season_weeks, weekly_rainfall[6:46], label='data')

rainfall_curve = np.polyfit(rainy_season_weeks, weekly_rainfall_season, 2)
p = np.poly1d(rainfall_curve)
plt.plot(rainy_season_weeks, p(rainy_season_weeks), label='fitted curve')
plt.legend();
```

```python
# formula for fitted curve
print('p = \n', p)
```

```python
# formula for derivative of fitted curve
print('p\' = ', round(rainfall_curve[0]*2,4), 'p  + ', round(rainfall_curve[1],4))
```

In order to take a look at the rates of change we will plug the weekly rainfall figures into the derivative equation shown above.  They should be positive up until around week 25 or 26 and they they should be negative.

```python
def p_prime(x):
    y = rainfall_curve[0]*2*x + rainfall_curve[1]
    return y
```

```python
# use list comprehension to permit multiplication by a float
p_prime_list = [round(p_prime(i),3) for i in rainy_season_weeks]
```

```python
# list should initially show positive values, then hit zero around the midpoint and then begin to show negative values
print(p_prime_list)
```
